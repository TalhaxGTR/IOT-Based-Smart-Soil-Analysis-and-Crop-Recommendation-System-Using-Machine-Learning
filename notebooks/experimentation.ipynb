# Smart Soil Analysis — Experimentation Notebook

This notebook contains exploratory data analysis and model experimentation.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
crop_df = pd.read_csv('../dataset/Crop_recommendation.csv')
fert_df = pd.read_csv('../dataset/fertilizer_recommendation_dataset.csv')

print('Crop dataset shape:', crop_df.shape)
print('Fertilizer dataset shape:', fert_df.shape)
```

```python
# Crop label distribution
crop_df['label'].value_counts().plot(kind='bar', figsize=(14,5), title='Crop Distribution')
plt.tight_layout()
plt.show()
```

```python
# Feature correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(crop_df.drop('label', axis=1).corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation — Crop Dataset')
plt.show()
```
